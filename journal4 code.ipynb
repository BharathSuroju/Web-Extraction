```python
import os, glob
import re
T = open('fileFinal5.txt',mode = 'r',encoding="utf-8")
U = open('fileFinal6.txt',mode = 'w',encoding="utf-8")
alltext = T.read()
pattern = re.compile('.*[(]\d{4}.*?[)].*\.$',re.MULTILINE)
match = pattern.finditer(str(alltext))
for i in match:
    U.write(str(i.group(0)))
    U.write('\n')
T.close()
U.close()
```

```python
import os, glob
import re
T = open('fileFinal.txt',mode = 'r',encoding="utf-8")
U = open('fileFinal3.txt',mode = 'w',encoding="utf-8")
alltext = T.read()
pattern = re.compile(r'(.*[(]\d{4}.*?[)].*)([(]\d{4}.*?[)])',re.DOTALL)
match = pattern.finditer(str(alltext))
for i in match:
    print(i.group(1))
T.close()
U.close()
```

```python
from pybtex.database.input import bibtex
import xlsxwriter
from xlsxwriter import Workbook
import os, glob,xlwt
wbk = Workbook('Bharath_journal4.xlsx')
sheet = wbk.add_worksheet('data')
import os, glob,xlwt
row = 0
author =''
parser = bibtex.Parser()
list_of_files = glob.glob(r'C:/bharath/5731/some/bib downloads/*.bib') 
for file in list_of_files:
    print(file)
    bib_data = parser.parse_file(file)
    list1 = bib_data.entries.keys()
    for i in list1:
        print(i,file)
        journal = bib_data.entries[i].fields['journal']
        title = bib_data.entries[i].fields['title']
        year = bib_data.entries[i].fields['year']
        author_list = bib_data.entries[i].persons['author']
        if author_list is None:
            pass
        else:
            for j in range(0,len(author_list)):
                author = author + str(bib_data.entries[i].persons['author'][j])    
        sheet.write(row,0,journal)
        sheet.write(row,1,title)
        sheet.write(row,2,year)
        sheet.write(row,3,author)
        row+=1
        author = ''
wbk.close()
```
